# Factor Groups
For this page I'll display the factor group generated by each normal subgroup.

```python
import sys  
sys.path.append('../')
    
from groupstartup import f, elements, e, subgroups, normal_subgroups
```

```python
allsets = {}
for subgroup in normal_subgroups:
    sets = [{f(element, subgroupelement) for subgroupelement in subgroup} for element in elements]
    uniquesets = []
    for _set in sets:
        if not _set in uniquesets:
            uniquesets.append(_set)
    sets = [list(_set) for _set in uniquesets]
    allsets[''.join(subgroup)] = sets
    print('Factor group generated by ' + ', '.join(subgroup)+':')
    print('{['+'], ['.join([', '.join(_set) for _set in sets])+']}')
```

I didn't find anything of particular note here, although it's somewhat interesting that elements that usually are grouped together(ibf7, 4cdg) are split up in some factor groups. Note that the sets in the factor group generated by the center are the same as the sets corresponding to specific inner automorphisms as determined in page 8. This should be expected since the factor group generated by the center is isomoporphic to the group of inner automorphisms.
